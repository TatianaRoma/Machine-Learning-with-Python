####12) Add columns bottom10  and top10

```python
min = np.min(df['Premature_death_Value'])
max = np.max(df['Premature_death_Value'])
range = (max - min)/10
```

```python
df['bottom10'] = [1 if d <= min + range else 0 for d in df['Premature_death_Value']]
```

```python
df['top10'] = [1 if d >= max - range else 0 for d in df['Premature_death_Value']]
```

```python
#only 2 couties in the top 10%
df[df.top10 == 1]
```

```python
# 149 counties in the bottom 10 %
len(df[df.bottom10 == 1])
```

####17) Sort according the years of life lost by state. Where is Maine in the ranking ?

#####"Premature death is represented by the years of potential life lost before age 75 (YPLL-75). Every death occurring before the age of 75 contributes to the total number of years of potential life lost. The YPLL measure is presented as a rate per 100,000 population and is age-adjusted to the 2000 U.S. population." 
The above description is copied from the dataset's documentation. Based on it I can assume that since the value is already represented as a rate per 100000 people, the size of the state or the county does not matter. So I can get the mean of county rates to represent the state rate of premature death.

```python
x = df[['State',"Premature_death_Value"]].groupby('State')
y = x[['Premature_death_Value']].mean().sort('Premature_death_Value', ascending = False)
y.rename(columns = {'Premature_death_Value': 'Years_Of_Life_Lost_Per_State'}).reset_index()
#x.ix[x[['Premature_death_Value']].transform(sum).sort('Premature_death_Value').index]
```

Maine is number 36 in this ranking (0 based indexing).
